```python
X_train_60 = load_json('../../data/X_train_60_2.json')
```

```python
y_train_60 = load_json('../../data/y_train_60_2.json')
```

```python
X_dev = load_json('../../data/X_dev_2.json')
```

```python
y_dev = load_json('../../data/y_dev_2.json')
```

```python
X_train_60_split_xpos = []

for sentence in X_train_60:
    new_sentence = []

    for word in sentence:
        new_word = {}
        
        for feature in word:
            if feature == 'xpos':
                new_word.update(split_xpos(word[feature], 'word'))
            elif feature == 'prev_word_xpos':
                new_word.update(split_xpos(word[feature], 'prev_word'))
            elif feature == 'word_before_prev_word_xpos':
                new_word.update(split_xpos(word[feature], 'word_before_prev_word'))
            elif feature == 'next_word_xpos':
                new_word.update(split_xpos(word[feature], 'next_word'))
            elif feature == 'word_after_next_word_xpos':
                new_word.update(split_xpos(word[feature], 'word_after_next_word'))
            else:
                new_word[feature] = word[feature]
    
        new_sentence.append(new_word)
    
    X_train_60_split_xpos.append(new_sentence)
```

```python
verify_prepped_data(X_train_60_split_xpos, y_train_60)
```

```python
X_dev_split_xpos = []

for sentence in X_dev:
    new_sentence = []

    for word in sentence:
        new_word = {}
        
        for feature in word:
            if feature == 'xpos':
                new_word.update(split_xpos(word[feature], 'word'))
            elif feature == 'prev_word_xpos':
                new_word.update(split_xpos(word[feature], 'prev_word'))
            elif feature == 'word_before_prev_word_xpos':
                new_word.update(split_xpos(word[feature], 'word_before_prev_word'))
            elif feature == 'next_word_xpos':
                new_word.update(split_xpos(word[feature], 'next_word'))
            elif feature == 'word_after_next_word_xpos':
                new_word.update(split_xpos(word[feature], 'word_after_next_word'))
            else:
                new_word[feature] = word[feature]
    
        new_sentence.append(new_word)
    
    X_dev_split_xpos.append(new_sentence)
```

```python
verify_prepped_data(X_dev_split_xpos, y_dev)
```

```python
X_dev_final = [[{feature:word[feature] for feature in word if feature not in ('contains_relative_pronoun_before', 'pronoun_tag', 'contains_conjunction_before', 'conjunction_tag', 'count_of_verbs_before')} for word in sentence] for sentence in X_dev_split_xpos]
```

```python
%%time
crf = sklearn_crfsuite.CRF(
    algorithm='lbfgs',
    c1=0.1,
    c2=0.1,
    max_iterations=200,
    all_possible_transitions=True,
    linesearch='MoreThuente',
    num_memories=8,
    period=12 
)
crf.fit(X_train_60_final, y_train_60)
```

```python
labels = list(crf.classes_)
y_pred = crf.predict(X_dev_final)
print(metrics.flat_f1_score(y_dev, y_pred, average='weighted', labels=labels))
print(metrics.flat_classification_report(y_dev, y_pred, labels=labels, digits=3))
labels.remove('')
print(metrics.flat_f1_score(y_dev, y_pred, average='weighted', labels=labels))
print(metrics.flat_classification_report(y_dev, y_pred, labels=labels, digits=3))
punctuated_sentences = punctuate(X_dev_final, y_pred)
print(punctuated_sentences[:100])
```
