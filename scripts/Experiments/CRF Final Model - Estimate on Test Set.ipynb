### Estimate Pretrained Model on Test Set

```python
pkl_filename = '../../models/crf_model.pkl'
with open(pkl_filename, 'rb') as file:
    clf = pickle.load(file)
```

```python
labels = list(clf.classes_)
y_pred = clf.predict(X_test_20)
labels.remove('')
print(metrics.flat_classification_report(y_test_20, y_pred, labels=labels, digits=3))
punctuated_sentences = punctuate(X_test_20[:100], y_pred[:100])
print(punctuated_sentences)
```
