```python
from ipynb.fs.full.utilities import *
```

```python
X_train_60 = load_json('../../data/X_train_60_2.json')
```

```python
y_train_60 = load_json('../../data/y_train_60_2.json')
```

```python
X_dev = load_json('../../data/X_dev_2.json')
```

```python
y_dev = load_json('../../data/y_dev_2.json')
```

```python
X_dev_no_upos = [[{feature:word[feature] for feature in word if feature not in ('contains_relative_pronoun_before', 'pronoun_tag', 'contains_conjunction_before', 'conjunction_tag', 'count_of_verbs_before', 'upos', 'prev_word_upos', 'word_before_prev_word_upos', 'next_word_upos', 'word_after_next_word_upos')} for word in sentence] for sentence in X_dev]
```

```python
X_train_60_no_upos = [[{feature:word[feature] for feature in word if feature not in ('contains_relative_pronoun_before', 'pronoun_tag', 'contains_conjunction_before', 'conjunction_tag', 'count_of_verbs_before', 'upos', 'prev_word_upos', 'word_before_prev_word_upos', 'next_word_upos', 'word_after_next_word_upos')} for word in sentence] for sentence in X_train_60]
```

```python
%%time
crf = sklearn_crfsuite.CRF(
    algorithm='lbfgs',
    c1=0.1,
    c2=0.1,
    max_iterations=200,
    all_possible_transitions=True,
    linesearch='MoreThuente',
    num_memories=8,
    period=12 
)
crf.fit(X_train_60_no_upos, y_train_60)
```

```python
labels = list(crf.classes_)
y_pred = crf.predict(X_dev_no_upos)
print(metrics.flat_f1_score(y_dev, y_pred, average='weighted', labels=labels))
print(metrics.flat_classification_report(y_dev, y_pred, labels=labels, digits=3))
labels.remove('')
print(metrics.flat_f1_score(y_dev, y_pred, average='weighted', labels=labels))
print(metrics.flat_classification_report(y_dev, y_pred, labels=labels, digits=3))
punctuated_sentences = punctuate(X_dev_no_upos, y_pred)
print(punctuated_sentences[:100])
```
